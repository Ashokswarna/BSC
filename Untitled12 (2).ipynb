```python
def dice_coef(y_true, y_pred,df_for_loss):
    df_for_loss['agg_closing_stock'] = np.where(df_for_loss.agg_closing_stock> 0, 18.75*df_for_loss.agg_closing_stock, df_for_loss.agg_closing_stock)
    df_for_loss['y_pred']=y_pred
    df_for_loss['agg_closing_stock'] = np.where(df_for_loss.agg_closing_stock== 0, 37.5*df_for_loss.y_pred, df_for_loss.agg_closing_stock)
    return sum(abs(df_for_loss['agg_closing_stock']))
```

```python
def dice_loss(df_for_loss):
  def dice(y_true, y_pred):
    return -dice_coef(y_true,y_pred, df_for_loss)
  return dice
```
